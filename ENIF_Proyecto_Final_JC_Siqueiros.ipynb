```python
# Remplazar los valores numéricos con los valores categóricos proporcionados por los cátalogos a través de una funcion
def valores_categoricos(df, dictionary):
    for column in df.columns:
        column = column.lower()
        if column in dictionary:
            if "cve" in dictionary[column]:
                # Crear un diccionario a partir de las listas de cve y descrip
                cve_descrip_dict = {cve: descrip for cve, descrip in zip(dictionary[column]['cve'], dictionary[column]['descrip'])}
                df[column.upper()] = df[column.upper()].map(cve_descrip_dict)
    
    return df
```

```python
#Para "tmodulo" crear una columna de "rango de ingresos" usando una funcion
def rango_de_ingresos_tmodulo(df_tmodulo):
    # Usamos np.select para establecer la periodicidad en función de 'P3_8B'
    conditions = [
        (df_tmodulo['P3_8B'] == 'A la semana'),
        (df_tmodulo['P3_8B'] == 'Al mes'),
        (df_tmodulo['P3_8B'] == 'A la quincena'),
        (df_tmodulo['P3_8B'] == 'Al año')
    ]

    choices = [7, 30, 15, 365]

    periodicidad = np.select(conditions, choices, default=1)
    
    limites_ingresos = [0, 10000, 25000, 50000, np.inf]
    
    etiquetas = ['<10000', '10001-25000', '25001-50000', '>50000']

    # Calcula el rango de ingresos y agrega la columna al DataFrame
    ingresos_mensuales = round((df_tmodulo['P3_8A'] / periodicidad),2)*30
    
    df_tmodulo['Rango de ingresos'] = pd.cut(ingresos_mensuales, bins=limites_ingresos, labels=etiquetas)

    return df_tmodulo
```

```python
df_tmodulo = rango_de_ingresos_tmodulo(df_tmodulo)
df_tmodulo['Rango de ingresos']
```

```python
# Para "tmodulo" crear una columna de "rango de edad" usando una funcion
def rango_de_edad(df_tmodulo):
    limites_edad = [18, 25, 35, 50, 70, 98]
    etiquetas_grupo = ['18-25 años', '26-35 años', '36-50 años', '51-70 años', '71-98 años']
    df_tmodulo['Rango de Edad'] = pd.cut(df_tmodulo['EDAD'], bins=limites_edad, labels=etiquetas_grupo, right=False)
    return df_tmodulo
```

```python
df_tmodulo = rango_de_edad(df_tmodulo)
df_tmodulo['Rango de Edad']
```

## ¿Cuáles son las razones principales por las que adquirió cuentas bancarias?

```python
# Lista de columnas que contienen información sobre razones de adquisición de cuentas bancarias
columnas_razones = ['P5_8_01', 'P5_8_02', 'P5_8_03', 'P5_8_04', 'P5_8_05', 'P5_8_06', 'P5_8_07', 'P5_8_08', 'P5_8_09', 'P5_8_10']

# Selecciona las columnas correspondientes del DataFrame principal
df_razones = df_tmodulo[columnas_razones]

# Diccionario para almacenar las frecuencias de cada razón
razones_frecuencias = {}

# Itera sobre las columnas de razones
for columna in columnas_razones:
    # Calcula las frecuencias de cada valor en la columna
    frecuencias = df_razones[columna].value_counts()
    # Almacena las frecuencias en el diccionario de razones_frecuencias
    razones_frecuencias[columna] = frecuencias

# Concatena las frecuencias de todas las razones y suma las frecuencias por fila
razones_principales = pd.concat(razones_frecuencias.values(), axis=1).sum(axis=1)

# Ordena las razones por la cantidad de personas en orden descendente
razones_principales = razones_principales.sort_values(ascending=False)

# Configuración de la visualización
plt.figure(figsize=(12, 6))
sns.set_style('whitegrid')

# Utiliza una paleta de colores atractiva
colores = sns.color_palette("pastel", len(razones_principales))

# Creación del gráfico de barras
plt.bar(razones_principales.index, razones_principales.values, color=colores)

# Etiquetas y título del gráfico
plt.xlabel('Razones de Adquisición')
plt.ylabel('Cantidad de Personas')
plt.title('Razones Principales por las que se Adquirieron Cuentas Bancarias')
plt.xticks(rotation=90)

# Muestra la cantidad de personas en cada barra
for i, v in enumerate(razones_principales.values):
    plt.text(i, v, str(v), ha='center', va='bottom', fontsize=10, color='black')

```

## ¿Qué forma de pago utiliza con más frecuencia cuando realiza compras?

```python
# Seleccionar las columnas relevantes y crear una copia del DataFrame
df_preguntas_metodos_pago = df_tmodulo[
    [
        'P7_8_1_1', 'P7_8_1_2', 'P7_8_1_3',
        'P7_8_2_1', 'P7_8_2_2', 'P7_8_2_3',
        'P7_8_3_1', 'P7_8_3_2', 'P7_8_3_3',
        'P7_8_4_1', 'P7_8_4_2', 'P7_8_4_3'
    ]
].copy()

# Definir pares de columna y valores de reemplazo
col_val_pairs = [
    ('P7_8_1_1', 'Transferencia'),
    ('P7_8_1_2', 'Tarjeta'),
    ('P7_8_1_3', 'Efectivo'),
    ('P7_8_2_1', 'Transferencia'),
    ('P7_8_2_2', 'Tarjeta'),
    ('P7_8_2_3', 'Efectivo'),
    ('P7_8_3_1', 'Transferencia'),
    ('P7_8_3_2', 'Tarjeta'),
    ('P7_8_3_3', 'Efectivo'),
    ('P7_8_4_1', 'Transferencia'),
    ('P7_8_4_2', 'Tarjeta'),
    ('P7_8_4_3', 'Efectivo'),
]

# Función para reemplazar valores según los pares definidos
def replace_values(data_frame, col_val_pairs):
    for col, val in col_val_pairs:
        data_frame[col] = data_frame[col].replace({1: val, 2: np.nan})

# Aplicar la función de reemplazo
replace_values(df_preguntas_metodos_pago, col_val_pairs)

# Renombrar columnas según los pares definidos
df_preguntas_metodos_pago = df_preguntas_metodos_pago.rename(columns={col: val for col, val in col_val_pairs})

# Contar la cantidad de respuestas y resetear el índice
df_preguntas_metodos_pago = df_preguntas_metodos_pago.count().reset_index(name='count')

# Agrupar por el índice y sumar las cuentas
df_preguntas_metodos_pago = df_preguntas_metodos_pago.groupby('index')['count'].sum().reset_index(name='count')

# Crear un gráfico de barras horizontal para mostrar la distribución de forma de pago
plt.figure(figsize=(12, 6))
bars = plt.barh(df_preguntas_metodos_pago['index'], df_preguntas_metodos_pago['count'], color='red')
plt.xlabel('Cantidad de Personas')
plt.ylabel('Forma de Pago')
plt.title('Forma de Pago Utilizada')

plt.gca().invert_yaxis()  # Invertir el eje y para que las formas de pago más comunes estén arriba
```
